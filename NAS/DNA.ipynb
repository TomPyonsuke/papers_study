#Blockwisely Supervised Neural Architecture Search with Knowledge Distillation (DNA)#
https://arxiv.org/pdf/1911.13053.pdf
##Main Contribution##
Proposed a way to solve the inaccurate evaluatation of sub-networks by using the same weights learned from training over-parameterized super network.
##Search Space##
image.png

* mobilenet-v2 inverted bottleneck w/ squeeze & excitation
* 5 searchable blocks
* each block has pre-defined cells with fixed number of layers and channels (a total of 16 searchable cells)
* searchable kernel size and expansion ratio: {3,5,7}, {3,6}

##Methods##
###Train###
Use efficient-b7 as teacher model to perform block-wise knowledge distillaton. Specifically, use input and output feature maps of each block from the teacher network to train **over-parametrized cells** at different blocks.

image.png

###Search###
####Step 1####
Use DFS to build a lookup table for the prediction power of all architectures in the search space. Represent prediction power as the sum of relative l1 losses over all blocks.

**Relative 11 loss:**
$$
\mathcal{L}_R(x,y)=\frac{||x-y||_1}{\sigma(y)}
$$

To avoid duplicate forward computation, DFS is applied per block:
image.png

####Step 2####
Q: Given resource constraint, how to find architecture that maximizes prediction power?

A: Sort losses of all architectures per block, use greedy + backtracking to minimize total loss while making sure resource constraint is not broken.
